# **Python Implementation**

```r
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```r
file_path = '/content/Student_Performance.csv'
df = pd.read_csv(file_path)
df_subset = df.head(50)
output_file_path = 'subset_dataset.csv'
df_subset.to_csv(output_file_path, index=False)

# file_path = '/content/Advertising Budget and Sales.csv'
# df = pd.read_csv(file_path)
# df_subset = df.head(50)
# output_file_path = 'subset_dataset.csv'
# df_subset.to_csv(output_file_path, index=False)
```

```r
summary = df_subset.describe()
head = df_subset.head()

# Print the summary and the first few records
print("Summary of numeric columns for the first 50 rows:")
print(summary)

print("\nFirst few records of the subset dataset:")
print(head)
```

```r
df_subset=df_subset.drop('Extracurricular Activities', axis=1)
```

```r
df_subset
```

```r
X=df_subset.drop('Performance Index', axis=1)
```

```r
X.head()
```

```r
Y=df_subset['Performance Index']
Y.head()
```

```r
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```r
model = LinearRegression()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
mse = mean_squared_error(Y_test, predictions)
y_hat = model.predict(X_test)
r_squared_sklearn = r2_score(Y_test, y_hat)
print("Coefficient of Determination (R-squared) from scikit-learn:", r_squared_sklearn)
print(f'Mean Squared Error: {mse}')
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```r
x = df_subset['Hours Studied']
y = df_subset['Performance Index']
plt.scatter(x, y)
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
# Convert the data into arrays
x = np.array(x)
y = np.array(y)

# Calculate the coefficients
b1, b0 = np.polyfit(x, y, 1)

# Add the regression line
plt.plot(x, b1*x + b0, color='red')
plt.show()
```

```r
# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the independent variables
x = df_subset['Hours Studied']
y = df_subset['Sample Question Papers Practiced']
z = df_subset['Performance Index']

# Add the data points
ax.scatter(x, y, z)

# Fit a plane using np.linalg.lstsq
A = np.vstack([x, y, np.ones_like(x)]).T
plane_coef, _, _, _ = np.linalg.lstsq(A, z, rcond=None)

# Create a meshgrid for the plane
x_plane, y_plane = np.meshgrid(x, y)
z_plane = plane_coef[0] * x_plane + plane_coef[1] * y_plane + plane_coef[2]

# Add the regression plane
ax.plot_surface(x_plane, y_plane, z_plane, alpha=0.5)

# Add labels and title
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Sample Question Papers Practiced')
ax.set_zlabel('Performance Index')
plt.title('Multiple Linear Regression')

# Show the plot
plt.show()
```

```r
set.seed(101)  # Set a seed for reproducibility
split_index <- sample(1:nrow(df_subset), 0.7 * nrow(df_subset))
train_data <- df_subset[split_index, ]
test_data <- df_subset[-split_index, ]
```

```r
print(colnames(train_data))
```

```r
model <- lm(`Performance.Index` ~ ., data = train_data)

# # Rename the 'Performance Index' column
# colnames(train_data)[colnames(train_data) == 'Performance Index'] <- 'Performance_Index'

# # Build linear regression model
# model <- lm(Performance_Index ~ ., data = train_data)
```

```r
# Get the model residuals
model_residuals = model$residuals

# Plot the result
hist(model_residuals)
```

```r
# Plot the residuals
qqnorm(model_residuals)
# Plot the Q-Q line
qqline(model_residuals)
```

```r
# Making predictions on the test set
predictions <- predict(model, newdata = test_data)
```

```r
r_squared_r <- 1 - sum((predictions - test_data$`Performance Index`)^2) / sum((mean(test_data$`Performance Index`) - test_data$`Performance Index`)^2)
mse <- mean((predictions - test_data$`Performance Index`)^2)
summary(model)

cat("Coefficient of Determination (R-squared):", r_squared_r, "\n")
cat("Mean Squared Error:", mse, "\n")
cat("Coefficients:\n", coef(model), "\n")

```
